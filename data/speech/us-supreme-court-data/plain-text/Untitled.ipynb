```python

def is_special_char_above_A1(some_string):
    
    # iterate over each character
    for char in some_string:
        
        # check if its unicode numeric value is over A1
        if ord(char) > 161:
            return True
    
    # if we haven't returned yet, there
    # are no special characters
    return False
```
